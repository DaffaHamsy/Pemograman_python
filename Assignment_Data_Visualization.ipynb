```python
from google.colab import drive
import pandas as pd

# Mount Google Drive
drive.mount('/content/drive')

# Path ke file CSV
file_path = '/content/drive/My Drive/PSO/epl-goalScorer(20-21).csv'

# Baca file CSV
df = pd.read_csv(file_path)

# Mengonversi nama kolom menjadi huruf kecil
df.columns = df.columns.str.lower()

# Mengubah nama kolom 'player' menjadi 'player_name'
df.rename(columns={'player': 'player_name'}, inplace=True)

# 10 player terbanyak mencetak gol
top_scorers = df.nlargest(10, 'goals')
print("10 player terbanyak mencetak gol:")
print(top_scorers[['player_name', 'goals']])
print()

# 10 player terlama durasi bermain
top_minutes = df.nlargest(10, 'time')
print("10 player terlama durasi bermain:")
print(top_minutes[['player_name', 'time']])
print()

# 10 player terbanyak assist
top_assists = df.nlargest(10, 'assists')
print("10 player terbanyak assist:")
print(top_assists[['player_name', 'assists']])
print()

# 4 player pencetak gol terbanyak dari kesebelasan manchester city
man_city_top_scorers = df[df['team_title'] == 'Manchester City'].nlargest(4, 'goals')
print("4 player pencetak gol terbanyak dari kesebelasan Manchester City:")
print(man_city_top_scorers[['player_name', 'goals']])
print()

# Dari kesebelasan mana pencetak gol terbanyak
most_goals_squad = df.groupby('team_title')['goals'].sum().idxmax()
print("Pencetak gol terbanyak berasal dari kesebelasan:", most_goals_squad)
print()

# Berapa jumlah kartu kuning dan merah yang diterima pembuat assist terbanyak
top_assist_player = df.nlargest(1, 'assists')['player_name'].values[0]
assist_player_cards = df[df['player_name'] == top_assist_player][['yellow_cards', 'red_cards']].sum()
print("Jumlah kartu kuning dan merah yang diterima oleh pembuat assist terbanyak:")
print(assist_player_cards)
print()

# Siapa yang jumlah assist dan golnya paling banyak dan dari kesebelasan mana
most_goals_assists = df['goals'] + df['assists']
most_goals_assists_player = df.iloc[most_goals_assists.idxmax()]
print("Pemain dengan jumlah assist dan gol terbanyak:")
print(most_goals_assists_player[['player_name', 'team_title', 'goals', 'assists']])
print()

# 5 kesebelasan yang menyumbangkan gol terbanyak
top_scoring_teams = df.groupby('team_title')['goals'].sum().nlargest(5)
print("5 kesebelasan yang menyumbangkan gol terbanyak:")
print(top_scoring_teams)
print()

# Urutkan pemain berdasarkan jumlah gol terbanyak tanpa mendapatkan kartu sama sekali
df['total cards'] = df['yellow_cards'] + df['red_cards']
no_cards_top_scorers = df[df['total cards'] == 0].nlargest(10, 'goals')
print("Pemain dengan jumlah gol terbanyak tanpa mendapatkan kartu sama sekali:")
print(no_cards_top_scorers[['player_name', 'goals']])
```
